```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


"""This is script to find a fourth order polynomial fit to set of temperature vs 
time pairs. The input comes in the form of two columns TimeAve and TempAve. 
The code outputs a plot comparing the fit and data points"""


file_path = 'C:\\Users\\allen\\Downloads\\SrO2ReactorProfile.xlsx'
data = pd.read_excel(file_path)

cleaned_data = data.dropna(subset=['Time2:', 'Temperature2:'])
cleaned_time = cleaned_data['Time2:']
cleaned_temperature = cleaned_data['Temperature2:']

coefficients = np.polyfit(cleaned_time, cleaned_temperature, 4)
polynomial = np.poly1d(coefficients)
time_dense = np.linspace(cleaned_time.min(), cleaned_time.max(), 500)
temperature_polynomial_dense = polynomial(time_dense)

plt.figure(figsize=(10, 6))
plt.plot(cleaned_time, cleaned_temperature, 'o', label='Data Points')
plt.plot(time_dense, temperature_polynomial_dense, '-', label='4th Order Polynomial Fit')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xlabel('Time [Sec]')
plt.ylabel('Temperature [C]')
plt.legend()
plt.title('Experimental Temperature vs. Time and Polynomial Fit')
plt.show()

print("Polynomial Coefficients (from highest to lowest order):")
print(f"a: {coefficients[0]:.2e}")
print(f"b: {coefficients[1]:.2e}")
print(f"c: {coefficients[2]:.2e}")
print(f"d: {coefficients[3]:.2e}")
print(f"e: {coefficients[4]:.2e}")
```
